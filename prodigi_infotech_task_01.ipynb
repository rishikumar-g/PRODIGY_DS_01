# _Extraction of top 15 countries with highest population with respect to 2023:_


```python
total_population = df[df["Series Code"] == "SP.POP.TOTL"]

total_population_sorted_desc = total_population.sort_values(by="2022", ascending=False)
top_15_countries = total_population_sorted_desc.head(15)
print("Top 15 countries that are most populated according to 2023 :")
print(top_15_countries["Country Name"])
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.barplot(x="2022", y="Country Name", data=top_15_countries, palette="coolwarm")
plt.title(
    "Top 15 populated countries in 2022",
)
plt.xlabel("Population(in bn)", fontsize=10)
plt.ylabel("Country Name", fontsize=10)
plt.show()

plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 2)
sns.barplot(x="2001", y="Country Name", data=top_15_countries, palette="coolwarm")
plt.title("Top 15 populateed countries in 2001")
plt.xlabel("Population(in bn)", fontsize=10)
plt.ylabel("Country Name", fontsize=10)
plt.show()
```

```python
total_population_sorted_asc = total_population.sort_values("2022", ascending=True)
bottom_15_countries = total_population_sorted_asc.head(15)

print("Bottom 15 populated countries :")
print(bottom_15_countries["Country Name"])
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.barplot(x="2022", y="Country Name", data=bottom_15_countries, palette="coolwarm")
plt.title("Bottom 15 populated countries in 2022")
plt.xlabel("Population", fontsize=10)
plt.ylabel("country Nmae", fontsize=10)
plt.show()

plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 2)
sns.barplot(x="2001", y="Country Name", data=bottom_15_countries, palette="coolwarm")
plt.title("Bottom 15 populated countries in 2001")
plt.xlabel("Population", fontsize=10)
plt.ylabel("country Nmae", fontsize=10)
plt.show()
```

# _Extraction of top 15 countries with highest male population:_


```python
male_population = df[df["Series Code"] == "SP.POP.TOTL.MA.IN"]
male_population_sort = male_population.sort_values(by="2022", ascending=False)
male_population_highest = male_population_sort.head(15)
print(male_population_highest["Country Name"])
```

# _Extraction of top 15 countries with highest female population_


```python
female_population = df[df["Series Code"] == "SP.POP.TOTL.FE.IN"]
female_population_sort = female_population.sort_values(by="2022", ascending=False)
female_population_highest = female_population_sort.head(15)
print(female_population_highest["Country Name"])
```

# _Creating bar pot for male and female population_


```python
# bar plot for male population
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.barplot(
    x="2022", y="Country Name", data=male_population_highest, palette="coolwarm"
)
plt.xlabel("Male Population")
plt.ylabel("Countries")
plt.title("Top 15 countries with higest male population(2022)")
plt.show()

# bar plot for female population
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 2)
sns.barplot(
    x="2022", y="Country Name", data=female_population_highest, palette="coolwarm"
)
plt.title("Top 15 countries with highest female population(2022)")
plt.xlabel(" Female population")
plt.ylabel("Countries")
plt.show()
```

# _Extraction of top 15 countries with lowest male and female population_


```python
male_lowest_pop = male_population_sort.tail(15)
print("Countries with lowest male populatuion")
print(male_lowest_pop["Country Name"])
```

```python
female_lowest_population = female_population_sort.tail(15)
print("countries with lowest female poulation (2022)")
print(female_lowest_population["Country Name"])
```

```python
# Bar plot for male lowest population
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.barplot(x="2022", y="Country Name", data=male_lowest_pop, palette="viridis")
plt.xlabel("Male Population")
plt.ylabel("Country Name")
plt.title("Top countries with lowest male population (2022)")
plt.show()

# Bar plot for female lowest population

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 2)
sns.barplot(
    x="2022", y="Country Name", data=female_lowest_population, palette="viridis"
)
plt.xlabel(" Female Population")
plt.ylabel("Country Name")
plt.title("Top countries with lowest female population(2022)")
plt.show()
```

# _Top countries with male and female population:_


```python
merge_data = pd.merge(
    male_population, female_population, on="Country Name", suffixes=("_male", "_female")
)
merge_data.head()
```

```python
merge_data.tail()
```

```python
merge_data["Total population"] = merge_data["2022_male"] + merge_data["2022_female"]
```

```python
merge_data
```

```python
total_population_sorted = merge_data.sort_values(by="Total population", ascending=False)
total_population_sorted.head()
```

```python
top_15_countries_totpop = total_population_sorted.head(15)
top_15_countries_totpop
```

```python
bottom_15_countries_totpop = total_population_sorted.tail(15)
bottom_15_countries_totpop
```

```python
# Bar plot total population of world 2022

plt.figure(figsize=(18, 8))
sns.barplot(
    x="Country Name",
    y="2022_female",
    data=top_15_countries_totpop,
    bottom=top_15_countries_totpop["2022_male"],
    color="red",
    label="female population",
    width=0.75,
)
sns.barplot(
    x="Country Name",
    y="2022_male",
    data=top_15_countries_totpop,
    color="blue",
    label="Male population",
    width=0.75,
)

plt.title("Total population of top 15 contries (2022)")
plt.xlabel("Country Name")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```

# _Total populaiton of bottom 15 countries_


```python
plt.figure(figsize=(16, 8))
sns.barplot(
    x="Country Name",
    y="2022_male",
    data=bottom_15_countries_totpop,
    bottom=bottom_15_countries_totpop["2022_female"],
    color="red",
    label="Male population",
)
sns.barplot(
    x="Country Name",
    y="2022_female",
    data=bottom_15_countries_totpop,
    color="blue",
    label="Female population",
)
plt.xlabel("Country Name")
plt.ylabel("population")
plt.title("Total population of bottom 15 countries (2022)")
plt.xticks(rotation=45)
plt.show()
```
